Pytania biznesowe (po wyczyszczeniu)

```python
# Używając df_clean:
df_clean['wynagrodzenie'] = pd.to_numeric(df_clean['wynagrodzenie'], errors='coerce')
# Pytanie 1: Jakie jest średnie wynagrodzenie per dział?
# Wskazówka: df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2)
srednie_wyn = df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2)
print("Średnie wynagrodzenie per dział:")
print(srednie_wyn)
# Pytanie 2: Jaka jest średnia ocena per dział?
# Wskazówka: df_clean.groupby('dzial')['ocena_roczna'].mean().round(2)
srednia_ocena_dzial = df_clean.groupby('dzial')['ocena_roczna'].mean().round(2)
print("\nŚrednia ocena per dział:")
print(srednia_ocena_dzial)
# Pytanie 3: Kto jest najlepiej opłacany (top 3)?
# Wskazówka: df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial', 'wynagrodzenie']]
top_3_place = df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial', 'wynagrodzenie']]
print("\nTop 3 najlepiej opłacanych:")
print(top_3_place)
# Pytanie 4: Ilu pracowników jest w każdym dziale?
# Wskazówka: df_clean['dzial'].value_counts()
liczba_pracownikow = df_clean['dzial'].value_counts()
print("\nLiczba pracowników w każdym dziale:")
print(liczba_pracownikow)
```

Rozszerzenie biznesowe (dla szybkich)

```python
df_clean['data_zatrudnienia'] = pd.to_datetime(df_clean['data_zatrudnienia'], errors='coerce')
# Dodatkowe pytania dla kierownika HR:
# 1. Ile lat pracuje najdłuższy staż? (dzisiejsza data - najwcześniejsza data_zatrudnienia)
dzis = pd.to_datetime('today')
najdluzszy_staz = (dzis - df_clean['data_zatrudnienia'].min()).days // 365
print(f"Najdłuższy staż w firmie wynosi: {najdluzszy_staz} lat")
# 2. Kto ma najwyższą ocenę w każdym dziale? (groupby + idxmax)
najlepsi_w_dziale = df_clean.loc[df_clean.groupby('dzial')['ocena_roczna'].idxmax()]
print("\nPracownicy z najwyższą oceną w każdym dziale:")
print(najlepsi_w_dziale[['dzial', 'imie', 'ocena_roczna']])
# 3. Czy jest korelacja między stażem a wynagrodzeniem?
staz_dni = (dzis - df_clean['data_zatrudnienia']).dt.days
korelacja = staz_dni.corr(df_clean['wynagrodzenie'])
print(f"\nKorelacja między stażem a wynagrodzeniem: {korelacja.round(2)}")
```
