```python
import pandas as pd
import numpy as np
# Series z listy — domyślny indeks (0, 1, 2, ...)
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

Zadanie 1a: Utwórz Series 
pensje z dict:

```python
# Klucze: 'Anna', 'Jan', 'Ewa', 'Marek', 'Kasia'
# Wartości: 5500, 7200, 4800, 9100, 6300
dane_pensje = {'Anna': 5500, 'Jan': 7200, 'Ewa': 4800, 'Marek': 9100, 'Kasia': 6300}
pensje = pd.Series(dane_pensje)
print(pensje)
```

Zadanie 1b: Odpowiedz na pytania o 
pensje 

```python
# 1. Wyświetl imię osoby zarabiającej najwięcej (użyj idxmax())
najwiecej_imie = pensje.idxmax()
print(f"Najwięcej zarabia: {najwiecej_imie}")
# 2. Oblicz średnią pensję
srednia_pensja = pensje.mean()
print(f"Średnia pensja: {srednia_pensja}")
# 3. Wyświetl imiona osób zarabiających powyżej średniej (filtrowanie boolean)
powyzej_sredniej = pensje[pensje > srednia_pensja]
print("Osoby powyżej średniej:")
print(powyzej_sredniej)
# 4. Policz ile osób zarabia poniżej 6000 (użyj len() lub .sum())
ponizej_6000 = (pensje < 6000).sum()
print(f"Liczba osób zarabiających poniżej 6000: {ponizej_6000}")
```

Krok 2 — Operacje na Series
Zadanie 1c:

```python
# 1. Oblicz pensje po podwyżce 10% dla wszystkich (pensje * 1.10)
podwyzka = pensje * 1.10
# 2. Oblicz pensje po premii 500 zł dla wszystkich (pensje + 500)
premia = pensje + 500
# 3. Oblicz nową pensję = (pensja * 1.10) + 500
nowa_pensja = (pensje * 1.10) + 500
# 4. Wyświetl ranking nowych pensji od najwyższej (użyj sort_values)
ranking = nowa_pensja.sort_values(ascending=False)
print("Ranking nowych pensji:")
print(ranking)
```

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)
```

Zadanie 3a: Pierwszy rzut oka

```python
# 1. Wyświetl head(10)
# 2. Wyświetl shape
# 3. Uruchom info() — ile brakujących wartości widzisz?
print(penguins.head(10))
print(f"Kształt: {penguins.shape}")
print(penguins.info())
```

Zadanie 3b: Statystyki opisowe

```python
# 1. Uruchom describe() — co mówią ci średnie i odchylenia?
# 2. Uruchom describe(include='all') — co dodają kolumny tekstowe?
# 3. Jaka jest mediana masy ciała (body_mass_g)?
print(penguins.describe())
print(penguins.describe(include='all'))
mediana_masy = penguins['body_mass_g'].median()
print(f"Mediana masy ciała: {mediana_masy} g")
```

Zadanie 3c: Rozkłady wartości

```python
# 1. Ile jest pingwinów każdego gatunku? (value_counts na 'species')
print("Gatunki:")
print(penguins['species'].value_counts())
# 2. Ile pingwinów na każdej wyspie? (value_counts na 'island')
print("\nWyspy:")
print(penguins['island'].value_counts())
# 3. Rozkład płci (value_counts na 'sex') — czy jest zbalansowany?
print("\nPłeć:")
print(penguins['sex'].value_counts())
# 4. Jaki procent to gatunek Adelie? (normalize=True)
print("\nProcentowy udział gatunku Adelie:")
print(penguins['species'].value_counts(normalize=True)['Adelie'] * 100)
```

Zadanie 3d: Brakujące dane

```python
# 1. Policz brakujące wartości w każdej kolumnie (isna().sum())
braki = penguins.isna().sum()
print("Braki w kolumnach:")
print(braki)
# 2. Jaki procent danych brakuje w kolumnie 'sex'?
procent_brakow_sex = (penguins['sex'].isna().sum() / len(penguins)) * 100
print(f"Procent braków w kolumnie sex: {procent_brakow_sex:.2f}%")
```
