```python
import numpy as np
```

Krok 1 — Dane

```python
# Ceny 4 produktów w 3 sklepach
ceny = np.array([[29.99, 31.99, 28.99],
[49.99, 52.99, 47.99],
[14.99, 15.99, 13.99],
# Produkt A
# Produkt B
# Produkt C
[99.99, 104.99, 94.99]]) # Produkt D
sklepy = ['Centrum', 'Galeria', 'Online']
produkty = ['A', 'B', 'C', 'D']
print(f"Kształt macierzy: {ceny.shape}") # (4, 3)
```

Krok 2 — Zadania

```python
# Broadcasting: macierz (4×3) * skalar
ceny_brutto = ceny * 1.05
print(ceny_brutto)
```

```python
rabaty = np.array([0.10, 0.05, 0.15])
# Broadcasting: macierz (4×3) * wektor (3,)
ceny_rabat = ceny - rabaty
print(ceny_rabat)
```

```python
marza = np.array([[0.30], [0.40], [0.25], [0.50]])
# Broadcasting: macierz (4×3) * wektor kolumnowy (4×1)
koszt_zakupu = ceny / (1 + marza)
print(koszt_zakupu)
```

```python
# Podpowiedź: użyj .argmin(axis=1) i listy sklepów
najtansze_indeksy = ceny.argmin(axis=1)
for i, index in enumerate(najtansze_indeksy):
    print(f"Produkt {produkty[i]}: Najtańszy sklep to {sklepy[index]} ({ceny[i, index]})")
```
